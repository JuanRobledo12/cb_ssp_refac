#### Technical Costs

Se utiliza la función CB_FRACTION_CHANGE. Nidhi da esta explicación de la función:

This function calculates the costs adn benefits as a multiplier applied to the difference in a variable between two strategies, where that difference is defined by some change in a factor, e.g., km/l or tons-waste/person. frac_var gives the name of the variable that has the fraction. Invert tells us whether we need to flip the fractions ot make our calculation correct. If our effect variable is already in the denominator of our fraction (e.g., effect in L, fraction is km/L) then we are good. If the effect variable is in the numerator (e.g., effect in T, fraction is T/person) then we need to flip it. To be specific, let E_tx be the effect we observe (e.g., L of fuel) in the transformation let f_base and f_tx be the fractions of km/L in the base and transformed futures then the distance that has been traveled in the transformation, d_tx = E_tx*f_tx. traveling that same distance with the old efficiency would have required  E_base = d_tx/f_base L. So, E_tx-E_base = E_tx - d_tx/f_base = E_tx - Etx*f_tx/f_base  = E_tx(1-ftx/f_base)

```python
from typing import List
def cb_get_data_from_wide_to_long(data : pd.DataFrame, 
                                  strategy_code : List[str], 
                                  variables : List[str]
                                  ) -> pd.DataFrame:

    #global SSP_GLOBAL_LOG_VARIABLE_SEARCH
    SSP_GLOBAL_SIMULATION_IDENTIFIERS = ['primary_id', 'region', 'time_period']

    #if SSP_GLOBAL_LOG_VARIABLE_SEARCH:
    #    SSP_GLOBAL_LOG_OF_SEARCHED_VARS = SSP_GLOBAL_LOG_OF_SEARCHED_VARS.append(variables)
    
    if not isinstance(variables, list):
        variables = [variables]

    if not isinstance(strategy_code, list):
        strategy_code = [strategy_code]

    data_wide = data[data["primary_id"].isin(strategy_code)][SSP_GLOBAL_SIMULATION_IDENTIFIERS + variables].reset_index(drop = True)
    
    data_long = data_wide.melt(id_vars=SSP_GLOBAL_SIMULATION_IDENTIFIERS)
  

    return data_long  

```

```python
## Listas de variables de diferencia para cada gas
diff_vars_gases = {
    "n2o" : ["emission_co2e_n2o_ippu_production_plastic", "emission_co2e_n2o_ippu_production_paper", "emission_co2e_n2o_ippu_production_chemicals", "emission_co2e_n2o_ippu_production_other_product_manufacturing"],
    "hfcs" : ["emission_co2e_hfcs_ippu_production_chemicals", "emission_co2e_hfcs_ippu_product_use_product_use_ods_refrigeration", "emission_co2e_hfcs_ippu_production_metals", "emission_co2e_hfcs_ippu_product_use_product_use_ods_other", "emission_co2e_hfcs_ippu_production_electronics"],
    "pfcs" : ["emission_co2e_pfcs_ippu_production_electronics", "emission_co2e_pfcs_ippu_product_use_product_use_ods_other", "emission_co2e_pfcs_ippu_production_other_product_manufacturing", "emission_co2e_pfcs_ippu_production_chemicals"],
    "other_fgases" : ["emission_co2e_other_fcs_ippu_production_electronics", "emission_co2e_other_fcs_ippu_production_chemicals", "emission_co2e_other_fcs_ippu_product_use_product_use_ods_other"],

}

## Juntaremos dichos gases para cada estrategia
def build_df_gases_strat(strategia : int)-> pd.DataFrame:
    df_strat = []
    for k,v in diff_vars_gases.items():
        for i in v:
            df_parcial_gas = cb_get_data_from_wide_to_long(ssp_salidas, strategia, i)
            df_parcial_gas["gas"] = k
            df_strat.append(df_parcial_gas)
    
    return pd.concat(df_strat, ignore_index = True)

df_gases_strat = {
    "base" : build_df_gases_strat(0),
    "constrained" : build_df_gases_strat("PFLO:CONSTRAINED"),
    "unconstrained" : build_df_gases_strat("PFLO:UNCONSTRAINED"),
    "net_zero" : build_df_gases_strat("PFLO:NET_ZERO")    
}

def plot_gases_strat_compara(estrategia_compara : str, gas:str)-> None:
    df_gases_strat_base = df_gases_strat["base"].query(f"gas=='{gas}'").groupby(["primary_id", "time_period"]).agg({"value" : "sum"}).reset_index()
    df_gases_strat_tx = df_gases_strat[estrategia_compara].query(f"gas=='{gas}'").groupby(["primary_id", "time_period"]).agg({"value" : "sum"}).reset_index()

    df_gases_compara = pd.concat([df_gases_strat_base, df_gases_strat_tx], ignore_index = True)
    df_gases_compara.pivot(index = "time_period", columns = "primary_id", values = "value").plot(title = f"BASE vs {estrategia_compara}\nGAS : {gas}")
    
```

```python
plot_gases_strat_compara("constrained", "other_fgases")
```

```python
plot_gases_strat_compara("unconstrained", "other_fgases")
```

```python
plot_gases_strat_compara("net_zero", "other_fgases")
```

### Compara razones entre estrategias


```python

def agrupa_dfs(estrategia : str, *args)-> pd.DataFrame:

    df_compara = pd.concat(
        [i.set_index("time_period")[[estrategia]] for i in args],
        axis = 1
    )

    df_compara.columns = ["consumer_savings", "technical_cost", "env_pollution"]
    
    return df_compara
    
agrupa_dfs("PFLO:CONSTRAINED", df_waso_consumer_savings, df_waso_technical_cost, df_waso_env_pollution).plot.area(title = "WASO PFLO:CONSTRAINED")
```

### Función cb_waso_reduce_consumer_facing_food_waste

La función obtiene para cada estrategia el consumer food waste generado en la simulación (variable qty_waso_total_food_produced_tonne) y construye un consumer food waste counterfactual que es igual al consumer food waste dividido entre el waso waste per capita scalar food. Se obtiene la diferencia entre el consumer food waste y el contrafactual para obtener el consumer food waste avoided. Por último, a esa diferencia se le multiplica el factor de costos de 100 ($100/ton)

```python
net_zero_waso_tech_cost = ssp_salidas.query("primary_id == 'PFLO:NET_ZERO'")[["time_period", "qty_waso_total_food_produced_tonne", "factor_waso_waste_per_capita_scalar_food"]]
net_zero_waso_tech_cost["consumer_food_waste"] = net_zero_waso_tech_cost["qty_waso_total_food_produced_tonne"]
net_zero_waso_tech_cost["consumer_food_waste_counterfactual"] = net_zero_waso_tech_cost["consumer_food_waste"]/net_zero_waso_tech_cost["factor_waso_waste_per_capita_scalar_food"]
net_zero_waso_tech_cost["consumer_food_waste_avoided"] = net_zero_waso_tech_cost["consumer_food_waste"] - net_zero_waso_tech_cost["consumer_food_waste_counterfactual"]
net_zero_waso_tech_cost["technical_costs"] = net_zero_waso_tech_cost["consumer_food_waste_avoided"]*10
net_zero_waso_tech_cost.set_index("time_period").drop(columns = ["qty_waso_total_food_produced_tonne", "factor_waso_waste_per_capita_scalar_food"]).plot(title = "NET ZERO\ncb:waso:technical_cost:consumer_food_waste:X\nMultiplicador de costo $10/ton")
```
